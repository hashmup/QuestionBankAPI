```python
import requests
from bs4 import BeautifulSoup
from pprint import pprint
```

```python
target_url = 'http://doors.stanford.edu/~sr/universities.html'
r = requests.get(target_url)         #requestsweb
soup = BeautifulSoup(r.text, 'lxml') #
```

```python
sql = "INSERT INTO schools (name, domain) VALUES "
for li in soup.find('ol').find_all('li'):
    name = li.find('a').get_text().replace('\n','').replace('\t', '').lstrip()
    domain = li.find(text=True, recursive=False).replace('\n', '').replace('(','').replace(')', '')
    if not name or not domain:
        continue
    sql += '("' + name + '","' + domain + '"),'
print(sql[:-1])
        
#     print(li.find('a')['value'], li['value'])
```
